# Lets change the datetime column last_ts to integer

## Considering we are calculating the last transaction from 16th of February 2023, lets calculate the number of days since the last transaction 

```python
# specify date
specified_date = pd.Timestamp('2023-02-16')

# calculate difference in days
df['days_since_last_txn'] = (specified_date - df['last_ts']).dt.days
```

### Setting 0 to all Nan values in  the fields related to Received transactions if recv_txn is zero

```python
recv_na_cols = ['min_val_recv', 'max_val_recv', 'avg_val_recv', 'sd2_val_recv', 'days_diff_first_last_recv']
df.loc[df['recv_txn'] == 0, recv_na_cols] = df.loc[df['recv_txn'] == 0,recv_na_cols].fillna(0)
```

```python
df.head()
```

### Setting 0 to all Nan values in  the fields related to Sent transactions if sent_txn is zero

```python
sent_na_cols = ['min_val_sent', 'max_val_sent', 'avg_val_sent', 'sd2_val_sent', 'days_diff_first_last_sent']
df.loc[df['sent_txn'] == 0, sent_na_cols] = df.loc[df['sent_txn'] == 0, sent_na_cols].fillna(0)
```

### Setting 0 to all Nan values in  the fields related to Contract transactions if to_contract_txn is zero

```python
contract_na_cols = ['tot_val_to_contract','min_val_to_contract', 'max_val_to_contract', 'avg_val_to_contract']
df.loc[df['to_contract_txn'] == 0, contract_na_cols] = df.loc[df['to_contract_txn'] == 0, contract_na_cols].fillna(0)
```

### Setting 0 to all Nan values in  the fields related to ERC20 transactions if erc20_txn is zero

```python
df.loc[df['erc20_txn'] == 0, 'sum_val_erc20'] = df.loc[df['erc20_txn'] == 0, 'sum_val_erc20'].fillna(0)
```

### Setting 0 to all Nan values in  the fields related to ERC721 transactions if erc721_txn is zero

```python
df.loc[df['erc721_txn'] == 0, 'sum_val_erc721'] = df.loc[df['erc721_txn'] == 0, 'sum_val_erc721'].fillna(0)
```
